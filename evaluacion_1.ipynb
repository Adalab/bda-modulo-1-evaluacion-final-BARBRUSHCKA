```python
def actualizar_stock(nombre, cantidad_agregar):
    for producto in inventory:
        if producto ["nombre"]==nombre:
            producto ["cantidad"]+= cantidad_agregar
            print(f"stock actualizado -≻{producto["nombre"]}:{producto["cantidad"]}")
            return

        print("producto no encontrado")   
            
```
